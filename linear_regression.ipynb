```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
def MSE(Y, Y_hat):
    N = Y.shape[0]
    MSE = 1/(2*N) * np.mean(np.power(Y - Y_hat, 2))
    return MSE
```

```python
class LinearRegression:
    
    def __init__(self, data, lr):
        self.X, self.Y = data
        self.Y = self.Y.reshape(-1,1)
#         self.X = np.hstack(np.ones_like(self.X[1]))
#         temp = self.Y.copy()
#         temp = np.hstack([self.Y, np.array([0])])
#         print(self.Y.shape)
        self.w = np.zeros_like(self.Y).reshape(-1,1)
        self.N = self.X.shape[0]
        
        #print(self.X.shape, self.w.shape, self.Y.shape)
        self.lr = lr
    
    def prediction(self):
        return self.X @ self.w
    
    
    def GD(self):
        pred = self.prediction()
        #print("PRED SHAPE ", pred.shape, self.Y.shape, self.X.shape)
        grad = 1/self.N * ( self.X.T @ (pred-self.Y))
        
        self.w = self.w - lr * grad
        return pred
    
    def fit(self, epochs):
        for i in range(epochs):
            pred = self.GD()
            print("Loss is ", MSE(pred, self.Y))
        
```

```python
data = make_regression()
data[0].shape, data[1].shape
```

```python
lr = 1e-2
model = LinearRegression(data, lr)
```

```python
P = model.prediction()
```

```python
model.fit(1000)
```
